```python
# EDA
```

```python
# sepal vs pedal 
```

```python
sns.scatterplot(data=data,x='petal_length',y='sepal_length',hue='species')
plt.title('Relationship Between Sepal Lenght & Petal lenght')
plt.xlabel('Petal Lenght')
plt.ylabel('Sepal_Length')
plt.show()
```

```python
sns.scatterplot(data=data, x='petal_width', y='sepal_width', hue='species', palette='Set2')
plt.title('Relationship Between Sepal Width & Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=data,x='sepal_length',y='sepal_width',hue='species')
plt.title('Relation Between Sepal Length & Sepal Width')
plt.xlabel('Sepal Lenght')
plt.ylabel('Sepal Width')
plt.show()
```

```python
sns.scatterplot(data=data,x='sepal_length',y='petal_width',hue='species')
plt.title('Relation Between Sepal Length & Petal Width')
plt.xlabel('Sepal Lenght')
plt.ylabel('Petal Width')
plt.show()
```

```python
sns.scatterplot(data=data,x='sepal_width',y='petal_width',hue='species')
plt.title('Relation Between Sepal Width & Petal Lenght')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.show()
```

```python
sns.scatterplot(data=data,x='petal_width',y='petal_length',hue='species')
plt.title('Relation Between Petal Width  & Petal Lenght')
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')
plt.show()
```
